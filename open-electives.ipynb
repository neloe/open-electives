```python
import json
from scrape_classes import get_classes
import pandas as pd

with open('csreqs.json', 'r') as f:
    allowed = json.load(f)
allowed = [c for c in allowed if c.startswith('44')]
with open('cscore.txt', 'r') as f:
    core = [l.strip() for l in f.readlines()]
allowed =set(allowed) - set(core)
allowed.remove('44101')
allowed.remove('44495')
electives = sorted(allowed)

def highlightFull(s, column):
    return ['background-color: red' if s.loc[column] <= 0 else '' for v in range(5)]

offerings = get_classes()
electives = [c for c in offerings if c[1] in allowed]
formatted = [(c[0], c[1], c[2].title(), ' '.join(c[3:5]), int(c[-1]) ) for c in electives]
mostopen = max(formatted, key=lambda x: x[-1])[-1]
df = pd.DataFrame(formatted, columns=['CRN', 'Crse#', 'Course Name', 'Time', 'Open Seats'])
df.sort_values(by='Open Seats',ascending=False).style.apply(highlightFull, column='Open Seats', axis=1).hide_index()
```
